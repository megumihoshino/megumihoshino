```python
#STARTING EXPRESSION WITH PYTHON
'''
ekspresi pda pemrograman python artinya mengombinasikan satu atau lebih variable,
konstanta, operator, dan fugsi yg bermakna utk menghasilkan suatu nilai dlm
suatu tipe tertentu.
struktur umum: <operan1><operator><operan2>
ket:
1. operan berupa nilai, variable,konstanta, atau ekspresi lain
2. operator, fungsi std dlm mtk, berupa jumlah, kurang, kali, bagi, dsb
'''
#EXAMPLE
x = 15
y = 23
result = x+y
print(result)

#penggabungan dan replikasi pada list
angka = [1,2,3,4,5,6]
huruf = ['A','B','C','D','E','F']
gabung = angka+huruf
print(gabung)

#replikasi list dgn operator kali (*)
replikasi = huruf*3
print(replikasi)
```

```python
#JENIS-JENIS EKSPRESI
'''
1. ekspresi biner
memiliki dua operan, bisa meliputi operan jml, kali, bagi, pangkat, dsb
cth: x+y, x*y, a/b, a^2
2. ekspresi uner
memiliki btk dasar operasi dgn satu operan
cth: x+=1 (increment), x-=1 (decrement), negasi (not x)
'''

#penerapan ekspresi uner
x = False #var x bernilai false
x = not x #ekspresi not x, brrti outputnya not False atau True
print(x)

y = 8 #ini tuh mksdnya kyk y = y+1, brrti y = 8+1 = 9
y +=1
print(y)

z = 7 #sm kyk yg atas, z = z-1, brrti z = 7-1 = 6
z -=1
print(z)

p = 23
print(-p)

#ekspresi relasional
print(3>10)
```

```python
#JENIS-JENIS OPERATOR
'''
1. Operator aritmetika
<numerik><operator><numerik> = <numerik>
'''
a = 13
b = 5
print(a+b)
print(a*b)
print(a-b)
print(a/b)
print(a//b) #pembagian bulat
print(a%b) #sisa hasil pembagian
print(a**b) #memangkatkan nilai dari dua operan

'''
2. Operator relasional
<numerik><operator><numerik> = <boolean>
'''
p = 13
q = 10
print(p==q) #operator sama dengan, hasil true klo dua operan bernilai sm
print(p!=q) #hasil true klo operan tdk bernilai sama
print(p>q)
print(p<q)
print(p>=q)
print(p<=q)
```

```python
#PADA STRING, KITA MEMFOKUSKAN PADA UNICODE
#misal: a memiliki unicode 1, dan b memiliki unicode 2
#kalau ada operator a>b, itu false, krn unicode a dibwh unicode b
#klo a<b, itu true, krn unicode a dibwh unicode b

#EXAMPLE
a = "NEKOMA"
b = "KARASUNO"
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

```python
#OPERATOR LOGIKA
'''
ekspresi logika:
<boolean><operator><boolean> = <boolean>
1. AND atau &
logika hny akn menghasilkan true jika kedua operan bernilai true
cth: a and b = false, a&b = false
2. OR atau |
logika yg menghasilkan true jika salah 1 dr kedua operan bernilai true
cth: a or b = true, a|b = true
3. NOT
logika yg ber7an utk meembalikkan nilai logika dan operannya
cth: not a = true

other resources:
https://www.geeksforgeeks.org/python-logical-operators/
'''

#1. EXAMPLE AND
print(True and True)
print(True and False)
print(False and True)
print(False and False)

#2. EXAMPLE OR
print(True or True)
print(True or False)
print(False or True)
print(False or False)

#3. EXAMPLE NOT
print(not True)
print(not False)
```

```python
#4. OPERATOR ASSIGNMENT
'''
bertujuan utk melakukan proses assignment atau pemberian nilai pd suatu
variabel dg nilai tetap
operator assignment ini akn digunakan pd perulangan (sequences)
'''
#+=
a = 15
a += 7 #penyederhanaan dari operasi a = a+b
print(a)

a = 15
a -= 7 #penyederhanaan dari operasi a = a-b
print(a)

a = 15
a *= 7 #penyederhanaan dari operasi a = a*b
print(a)

a = 15
a /= 7 #penyederhanaan dari operasi a = a/b
print(a)

a = 15
a %= 7 #penyederhanaan dari operasi a = a%b
print(a)

#mungkin utk penjelasan lbh lanjut, krn yg di atas itu ekspresi uner
a = 15
a = a + 7  #ekspresi biner
print(a)

a = 15
a+=7 #ekspreso uner
print(a)

#keduanya akan menghasilkan nilai yg sama, yaitu 22
```
