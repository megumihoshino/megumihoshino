```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
xdata = [0.4, 0.6, 0.8]
ydata = [39.21, 30.48, 29.36]
plt.figure(figsize = (10,6))
plt.xlabel('Rapat Arus (J, mA/mm$^2$)', size = 'x-large')
plt.ylabel('Rata-Rata Ukuran Kristal (nm)', size = 'x-large')
plt.title('Rata-Rata Ukuran Kristal Lapisan Komposit Ni-AlN', size = 'xx-large')
plt.plot(xdata, ydata, color ='#ff0090', label= 'ydata', marker = 'o')
plt.legend(['Ni-AlN'], shadow = True)
plt.grid()
plt.show()
plt.savefig('grafik_ukurankristal_ratarata.png', dpi = 600, bbox_inches ='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
adata = [0.4, 0.6, 0.8]
bdata = [61.12, 76.18, 89.18]
cdata = [10.13, 8.66, 1.79]
edata = [28.75, 15.16, 9.03]
plt.figure(figsize = (10,6))
plt.xlabel('Rapat Arus (J, mA/mm$^2$)', size = 'x-large')
plt.ylabel('Komposisi Jumlah Atom Unsur (at.%)', size = 'x-large')
plt.title('Persentase Komposisi Atom Unsur Terhadap Rapat Arus Elektrodeposisi', size = 'xx-large')
plt.plot(adata, bdata, color ='#ff4500', label= 'bdata', marker = '^')
plt.plot(adata, cdata, color ='#b0bf1a', label= 'cdata', marker = 'v')
plt.plot(adata, edata, color = 'g', label = 'edata', marker = 'o')
plt.legend(['Ni', 'Al', 'N'], shadow = True)
plt.grid()
plt.show()
plt.savefig('grafik_ukurankristal_ratarata.png', dpi = 600, bbox_inches ='tight')
```
