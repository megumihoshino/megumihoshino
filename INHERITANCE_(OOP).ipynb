```python
#example1 (kelas induk lingkaran dan kls anak bola)

import math
class circle:
  def __init__(self, radius):
    self.radius = radius

  def find_area(self):
    return self.radius**2*math.pi


class sphere_area(circle): #kelas turunan dari kelas induk circle (buat kelas baru)
  def sphere(self):
      return self.radius**2*math.pi*4

#kelas luas lingkaran
circle_1 = circle(5)
area = circle_1.find_area()
print(f"area of circle: {round(area, 3)}")

#kelas luas bola
circle_2 = sphere_area(5)
area_new = circle_2.sphere()
print(f"area of sphere: {round(area_new, 3)}")
```

```python
#Override (mekanisme pewarisan)
'''
ketika kita membuat metode baru di kelas turunan (kls baru) dg nama yg sama
sprt metode di kls induk, itu akan menyebabkan metode baru menimpa (override)
metode dari kelas induk.'''
```

```python
import math
class circle:
  def __init__(self, radius):
    self.radius = radius

  def find_area(self):
    return self.radius**2*math.pi

  def half_circle(self):  #half_circle() hny didefinisikan di kls circle. kls sphere dpt lngsg menggunakan tnp perlu definisi ulang
    return self.find_area()/2


class sphere_area(circle): #kelas turunan dari kelas induk circle (buat kelas baru)
  def sphere(self):
      return self.radius**2*math.pi*4

#kelas luas lingkaran
circle_1 = circle(5)
area = circle_1.find_area()
print(f"area of circle: {round(area, 3)}")

#kelas luas bola
circle_2 = sphere_area(6)
area_new = circle_2.sphere()
print(f"area of sphere: {round(area_new, 3)}")

#kelas luas setengah lingkaran
half_circle_area = circle_2.half_circle()
print(f"half circle area: {round(half_circle_area, 3)}")
```

```python
#SUPER
'''menggunkan metode atau atribut dr kelas induk, tp tdk ingin menuliskan semua kode.
memanfaatkan konsep ini utk menghindari kode berulang.'''

class haikyuu:
  def __init__(self, nama):
    self.nama = nama

  def cetak_point(self):
      raise NotImplementedError #membantu memastikan bhw metode tsb diimplementasikan di kls turunan, meningkatkan fleksibilitas dan maintainability kode

class karasuno(haikyuu):
  def __init__(self, nama, bola_in):
    super().__init__(nama)
    self.bola_in = bola_in

  def cetak_point(self):
    return "nice kill!"

class nekoma(haikyuu):
  def __init__(self, nama, bola_out):
    super().__init__(nama)
    self.bola_out =  bola_out

  def cetak_point(self):
    return "dont mind, dont mind!"

karasuno_1 = karasuno("tsukishima", 1)
nekoma_1 = nekoma("kuroo", 0)

print(f"{karasuno_1.nama} ({karasuno_1.bola_in}), karasuno tim berkata: {karasuno_1.cetak_point()}")
print(f"{nekoma_1.nama} ({nekoma_1.bola_out}), nekoma tim berkata: {nekoma_1.cetak_point()}")
```

```python
class Animal:
  def __init__(self, name, age, species):
    self.name = name
    self.age = age
    self.species = species

class Cat(Animal):

  def deskripsi(self):
    return f"{self.name} adalah kucing berjenis {self.species} yang sudah berumur {self.age} tahun"

  def suara(self):
    return "meow!"

myCat = Cat("Neko", 3, "Persian")
print(myCat.deskripsi())
print(myCat.suara())
```

```python
arr = [-4, 3, -9, 0, 4, 1]

positive_count = 0
negative_count = 0
zero_count = 0

# Hitung jumlah nilai positif, negatif, dan nol
for num in arr:
  if num > 0:
    positive_count += 1
  elif num < 0:
    negative_count += 1
  else:
    zero_count += 1

# Hitung total elemen
total_count = len(arr)

# Hitung proporsi dengan dua desimal
positive_prop = round(positive_count / total_count, 2)
negative_prop = round(negative_count / total_count, 2)
zero_prop = round(zero_count / total_count, 2)

# Cetak proporsi
print(f"Proportion of positive values: {positive_prop}")
print(f"Proportion of negative values: {negative_prop}")
print(f"Proportion of zeros: {zero_prop}")
```
