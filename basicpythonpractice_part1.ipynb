```python
"""
implementasi slicing, 7an utk mengambil data berdsrkan indeks rentang tertentu
pola slicing: sequence[start:stop:step]
start: indeks pertama yg mau diambil, sifatnya inklusif
stop: indeks terakhir yg mau diambil, sifatnya ekslusif
step: jmlh elemen yg dilewati a.k.a interval, defaultnya nilai step adlh 1
mksd inklusif atau ekslusif
example: rentang utk nilai 3>x>=8
brrti nilai x = 4,5,6,7,8
example 2 3>x>8
brrti nilai x = 4,5,6,7
"""

a = ["tsukishima", "hinata", "yamaguchi", "tobio", "yachi", "daichi", 'sugawara']
print(a[0:5:2])
print(a[2:])
print(a[:4])
```

```python
#set, kumpulan item yg tdk beraturan, gabs diindexing, pke kurung kurawal
a = {1, 2,3,3,5,6,8,9,100}
print(type(a))
print(a)

#intersection (irisan) dan union (penggabungan) dlm himpunan
b = {1,2,3,4,5,6,7,8}
c = {2,4,6,8,10}

union = b.union(c)
print("Union:", union)

intersection = b.intersection(c)
print("Intersection:", intersection)
```

```python
#TRANSFORMASI ANGKA, KARAKTER, DAN STRING
#1. upper dan lower case
kata = 'tsukishima'
kata = kata.upper()
print(kata)

kata1 = 'TSUKISHIMA'
kata1 = kata1.lower()
print(kata1)
```

```python
#2. awalan dan akhiran
#rstrip(), menghapus whitespace sblh kanan, atau menghapus kt yg tdk diinginkan sblh kanan
print("tsukishima"    .rstrip())

#lstrip(), menghapus whitespace sblh kiri, menghapus kt yg tdk diinginkan sblh kiri
print("      tsukishima".lstrip())

#strip(), menghapus whitespace di awal dan akhir
print(     "tsukishima"     .strip())

#menghapus kt yg tdk diinginkan dgn strip()
a = 'hinatahinatahinatatsukishimahinatahinatahinata'
print(a.strip("hinata"))

#METODE STARSWITH, utk suatu kata pada awal str, output berupa bool
print('tsukishima sayang megumi'.startswith('tsukishima'))
print('tsukishima sayang megumi'.endswith('kei'))
```

```python
#MEMISAH DAN MENGGABUNGKAN STRING
#gabungin str
print(' '.join(['tsuksihima','megumi'])) #single quotes sblm join berfungsi sbg penanda klo dikasi 'whitespace' ato jarak

#nambah delimiter laen
print('2709'.join(['tsuksihima','megumi'])) #pemisahnya 2709

#split str = dr gabung ke misah
print('tsuksihima megumi'.split())

#memisahkan setiap baris pd str multi line
print('''allo,
aku tsukishima megumi.
aku istrinya kei.
aku suka buah pir,
dan aku bukan ikan.'''.split('\n'))

#mengganti elemen string
string = "ayok nonton haikyuu di Jepang"
print(string.replace("Jepang","Indonesia"))
```

```python
#MENGECEK STRING
#isupper() akan mengembalikan nilai True jika semua huruf dalam string adalah huruf kapital (uppercase)
#islower(), Kebalikannya, islower() akan mengembalikan nilai True jika semua huruf dalam string adalah huruf kecil (lowercase).
string = 'TSUKISHIMA123'
string1 = '456'
string2 = ' '
string3 = 'Tsukishima'
print(string.isupper())
print(string.islower())
print(string.isalpha()) #jika semua alfabet
print(string.isalnum()) #jika semua alfabet numerik
print(string1.isdecimal()) #jika menunjukkan desimal
print(string2.isspace()) #memastikan str berupa whitespace
print(string3.istitle()) #mengembalikan nilai True jika string berisi huruf kapital pada setiap kata pertamanya. Jika tidak, nilai False dikembalikan.
```
