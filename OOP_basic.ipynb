```python
#atribute
'''
dalam python,ada 2 jns atribute kelas yg dpt dibagi, yaitu atribute kelas
dan atribute objek atau instance.'''

class car:
  #class atribute
  color = "blue"

car_1 = car()
print(car_1.color)

car_2 = car()
print(car_2.color)

#change class atribute
car.color = "yellow"

print(car_1.color)
print(car_2.color)
```

```python
#CLASS CONSTRUCTOR
'''
pembangun kelas/class constructor adlh sebuah fungsi khusus python  yg digunakan
utk menentukan nilai awal atau kondisi awal dari suatu kelas. kita dpt
melakukan proses instansiasi.'''

class cat:
  #atribute instance
  def __init__(self):
    self.cat_type = "tabby"

cat_1 = cat()
cat_2 = cat()

print(cat_1.cat_type)
print(cat_2.cat_type)

#changing atribute instance
cat_1.cat_type = "calico"

# showing cat type atribute
print(cat_1.cat_type) #yang diubah cuma cat_1, makanya jadi calico
print(cat_2.cat_type)
```

```python
#METHOD
'''
method di sini merupakan perilaku atau tindakan yg dpt dilakukan oleh objek
atau kelas. pd pembuatan method, tentu saja kita akan membuat fungsi dlm kelas itu sendiri.
ada tiga method dlm OOP:
1. metode dr object (object method)
2. metode scr statis(static method)
3. metode dr class (class method)

dua metode terakhir erat kaitannya dgn dekorator. apa itu dekorator? dekorator
adalah fungsi dlm python yg mengembalikan fungsi lain, diawali dgn sintaks @ di awal.'''
```

```python
#example1
def my_decorator(func): #def fungsi bernama my_decorator, menerima func sbg parameter
  def wrapper(): #def fungsi wrapper(), sbg "pembungkus" yg menambahkan perilaku sblm dan sesudah eksekusi
    print("before execute")
    func()
    print("after execute")
  return wrapper #mengembalikan return sbg fungsi wrapper sbg hasilnya. return menyebabkan fungsi wrapper berjalan

#dekorasi fungsi dgn dekorator
@my_decorator
def hajimemashite():
  print("Irasshaimaseee!")

#memanggil fungsi yg sudah didekorasi
hajimemashite()
```

```python
'''
- Decorator my_decorator menambahkan fungsionalitas "before execute" dan
"after execute" ke fungsi hajimemashite.
- Fungsi wrapper adalah fungsi baru yang dikembalikan oleh decorator.
- Ketika hajimemashite dipanggil, fungsi wrapper dijalankan.
- Fungsi wrapper menjalankan "before execute", kemudian hajimemashite,
dan terakhir "after execute".'''
```

```python
#METODE DARI OBJEK (OBJECT METHOD)
'''ciri khas dari parameter ini adanya parameter "self".'''
#example

class car:
  def __init__(self, color, merk, speed):
    self.color = color
    self.merk = merk
    self.speed = speed

  def add_speed(self):
    self.speed += 45

car_1 = car("blue", "pajero", 70)
print("before adding: ")
print(car_1.speed)

car_1.add_speed() #calling metode tmbh kecepatan


print("after adding: ")
print(car_1.speed)
```

```python
#example 3
class VB_national_match:
  def __init__(self, team, win_match, prefecture):
    self.team = team
    self.win_match = win_match
    self.prefecture = prefecture

  def match(self):
    if self.win_match >= 5:
      print("congrats team!")
    elif self.win_match == 3:
      print("next match again, keep up the good play!")
    else:
      print("don't mind, team. you did great!")

team_1 = VB_national_match("karasuno", 6, "Miyagi")
print(f"Team: {team_1.team}, Match: {team_1.win_match}")
team_1.match()

team_2 = VB_national_match("nekoma", 3, "Tokyo")
print(f"Team: {team_2.team}, Match: {team_2.win_match}")
team_2.match()

team_3 = VB_national_match("inarizaki", 2, "Hyogo")
print(f"Team: {team_3.team}, Match: {team_3.win_match}")
team_3.match()
```

```python
#METODE DARI KELAS (CLASS METHOD)
'''
jika objek metode identik dg parameter self yg merujuk pd objek, sedangkan
class method memerlukan sebuah prameter yg merujuk pd kelas'''

class car:
  def __init__(self, merk):
    self.merk = merk

  @classmethod
  def intro_car(cls):
    print("this is class method of car")

car.intro_car()
car_1 = car("pajero")
car_1.intro_car()
```

```python
#example class method utk luas lingkaran

#tidak memerlukan objek utk dipanggil, dlm kasus di bwh, tdk memerlukan
#objek circle_1 utk dipanggil sprt pada kodingan di bwhnya (object method)
import math
class area_of_circle:
  @classmethod #
  def find_area_circle(cls, radius):
    return math.pi*radius**2

radius = 6

area = area_of_circle.find_area_circle(radius)
print(f"area of circle: {round(area,3)}")
```

```python
#example object method utk luas lingkaran

import math
class area_of_circle:
  def __init__(self, radius): #object method krn tdk didekorasi dgn class method
    self.radius = radius

  def find_area_circle(self):
    return self.radius**2*math.pi

#create object circle
circle_1 = area_of_circle(6)

#menghitung luas lingkaran
area = circle_1.find_area_circle()

print(f"area of circle: {round(area, 3)}")
```

```python
class Animal:
  def __init__(self, name, age, species):
    self.name = name
    self.age = age
    self.species = species

class Cat(Animal):

  def deskripsi(self):
    return f"{self.name} adalah kucing berjenis {self.species} yang sudah berumur {self.age} tahun"

  def suara(self):
    return "meow!"

myCat = Cat("Neko", 3, "Persian")
print(myCat.deskripsi())
print(myCat.suara())
```
