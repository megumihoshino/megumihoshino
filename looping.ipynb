```python
#PERULANGAN/LOOPING
'''
digunakan agar tdk menggunakan sintaks yg berulang krn tdk efektif
ada beberapa statement utk melakukan perulangan.'''

#1. FOR, definite iteration
'''
artinya proses iterasi atau perulangan ketika jml perulangannya ditentukan
scr eksplisit seblmnya. formatnya:
for <var> in <iterable>:
  <statement(s)>'''
#example
var_list = [2,4,6,8,10,12,14,16]
for i in var_list:
  print(i)

#example 2
for i in range(10):
  print(i)

#range adlh fungsi bawaan python yg akn menghasilkan urutan bilangan
#dimulai dari indeks 0
#sintaks umum range(start, stop, step)
```

```python
for i in range(1,13,2):
  print(i)
```

```python
#2. WHILE, sintaks dlm python yg berisfat indefinite iteration.
'''artinya sebuah proses iterasi yg akn berhenti ketika memenuhi kondisi
tertentu'''
#example 1
counter = 1 #inisiasi nilai 1
while counter <=6:
  print(counter)
  counter +=1 #incremnt, hingga menyentuh angka 6
```

```python
#3. FOR BERSARANG (nested loop)
'''program ini akan melakukan perulangan luar dan perulangan dalam. perulangan
luar akn diproses terlbh dulu, lalu perulangan dalam.'''
#example
for i in range(1, 4): #outer loop
  for j in range(1,4): #inner loop
    print(i, j)
```

```python
#KONTROL PERULANGAN
for i in range(3):
  print("outer loop:", i)
  for j in range(8):
    print("inner loop:",j)
    if j == 1:
      break
```

```python
#example2
for huruf in 'Tsuki shima':
  if huruf == ' ':
    break
  print('Huruf saat ini: {}'.format(huruf)) #jika bertemu ' ',maka
  #program akn berhenti

#example3, using continue
for huruf in 'Tsuki shima':
  if huruf ==' ':
    continue #statement ini akan mengabaikan statement yg berada
    #antara continue hingga akhir blok kode
  print('Huruf saat ini: {}'.format(huruf))
```

```python
#else setelah for
angka = [2,3,5,6,7,8,11]

for i in angka:
  if i == 8:
    print("ada hehe")
    break
else:
  print("ga ada hehe")
```

```python
#else setelah while
'''blok statement else akan selallu dieksekusi saat conditiion pd while mjd
salah'''
#example
count = 2

while count < 7:
  print("tsukishima")
  count += 1
else:
  print("eksekusi blok else")
```

```python
#kalau pake break
count = 15
while count > 0:
  count = count-2
  if count == 5:
     break
  print(count)
else:
  print("setop") #program terhenti di angka 5, tp tdk tercetak hasil eksekusi statement else

```

```python
#LIST COMPREHENSION, membuat list bru berdsrkan list yg sudah ada
nilai = [2,3,4,5,6,7]
pangkat = []
for i in nilai:
  pangkat.append(i**2)
print(pangkat)
```

```python
#other example
nilai = [2,3,4,5,6,7]
pangkat = [i**2 for i in nilai]
print(pangkat)
```

```python
#ganjil
oddNumber = []
for i in range(0,502):
  if i%2 == 1:
    oddNumber.append(i)
print(oddNumber)

#genap
evenNumber = []
for i in range(0,502):
  if i%2 == 0:
    evenNumber.append(i)
print(evenNumber)
```

```python
#EXCEPTIONS HANDLING
'''bisa ditangani menggunakan try-except statement'''

#example1
a = 0
print(1/0) #outputnya ZeroDivisionError, artinya program eror krn tdk bs
#melakukan operasi aritmetika dgn angka nol.
```

```python
#example 2, dgn try-except
a = 0
try:
  print(1/a)
except ZeroDivisionError:
  print('tdk bsa dibagi dgn nol')
```

```python
#RAISE EXCEPTION
'''digunakan "secara sengaja" utk membuat kode program yg ingin kita batasi dgn kondisi tertentu'''
#example1

comment = -3
if comment < 0:
  raise ValueError("haikyuu itu bagus no debat")
else:
  for i in range(comment):
    print(i+1)
```
