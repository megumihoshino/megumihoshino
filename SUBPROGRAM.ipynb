```python
#SUBPROGRAM
'''
Serangkaian instruksi dirancang utk melakukan operasi yang sering digunakan
dlm suatu program. terdiri dari dua jns:
1. fungsi
blok kode yg dpt menerima input, pemrosesan, dan mengembalikan output. fungsi yg
dibuat dpt ditentukan utk mengembalikam tipe data int, str, dsb.
2. prosedur
deretan instruksi yg jelas keadaan awal (initial state) dan keadaan akhirnya
(final state). memiliki cakupan yg kecil dan terbatas.
'''
import math
def find_luas_lingkaran(r):
  luas_lingkaran = math.pi*r**2
  return luas_lingkaran

lingkaran_1 = find_luas_lingkaran(3)
print(round(lingkaran_1, 2))

lingkaran_2 = find_luas_lingkaran(5)
print(round(lingkaran_2, 2))
```

```python
#1. Positional-or-keyword, menggunakan positional ataupun keyword argument ketika memanggili fungsi
#example

def greeting(name, message):
  return "Konnichiwa, " +name+ "!" +message #positional

print(greeting("Tsukishima","O genki desu ka?"))
print(greeting(message="Konbanwa!", name = "Megumi")) #keyword
```

```python
#2. positional-only
def perkalian(x,y,/):
  return x*y

print(perkalian(9, 80))
```

```python
#3. keyword-only, menggunakan "*" (asterisk)
def greeting(*, name, message):
  return "Hola, " +name+ "!" +message
print(greeting(message="Konbanwa!", name= "Tsukishima"))
```

```python
#5. var-keyword (variadic keyword parameter)
'''
parameter ini dpt menampung jml keyword yg bervariasi saat pemanggilan fungsi.
menggunakan sintaks **kwargs yg berperan dictionary.'''

def info(**kwargs):
  informasi = ""
  for key, value in kwargs.items():
    informasi += key + ':' + value + ","
  return informasi
print(info(nama = "megumi", usia = "23", occupation = "cloud engineer", ttl= "jakarta", lulusan = "unj"))
```

```python
#GLOBAL VARIABLE
'''merupakan suatu variabel yg didefinisikan di luar fungsi atau blok kode dan
dpt diakses dari seluruh bag program.'''
#example

x = 70 #var x merup var global yg dpt diakses di seluruh program

def add(y):
  result = x*y
  return result
first_val = add(15)
print(first_val)
```

```python
#LOCAL VARIABLE
'''hanya pada suatu fungsi atau blok kode tertentu'''
def add(x,y):
  loc_var = 70
  result = x*y/loc_var
  return result

first_val = add(89,45)
print(round((first_val),2))
```
