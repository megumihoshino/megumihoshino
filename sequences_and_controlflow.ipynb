```python
#PYTHON INTERPRETER
'''
- kode pd python akan ditransformasi mjd kode yg mudah dimengerti mesin dg
menggunakan program compiler atau interpreter.
- compiler merup. program yg akan menerjemahkan bhs pemrograman mjd bhs mesin
sblm dijalankan & menghasilkan output
- interpreter, akn menejermahkan bhs python satu per satu dan menghasilkan
output scr langsung.
'''
#BLOCK CODE
'''
blok kode merujuk pada potongan kode program python yg dijaankan sbg satu
unit. blok kode dpt berupa modul, fungsi, kelas, control flow, dsb.
'''
#EXAMPLE
for i in range(6):
  print(i) #kode perulangan utk cetak angka 0 sampai 5

#CASE SENSITIVE
'''artinya python memperlakukan huruf besar dan kecil sbg karakter
berbeda dlm penamaan variable'''
#example
text = "Cloud"
Text = "Engineer"
print(text)
print(Text)
```

```python
#ONE LINER
'''
membuat kode hny dlm satu baris atau single statement. keunggulan dari
python dibanding dgn prog lang yg lainnya.'''

#menukar dua variable, blm pakai one liner
a = 5 #inisiasi a sbg 5
b = 7 #inisiasi b sbg 7
tukar = a #inisiasi var tukar dgn nilai a, mjd 5
a = b #inisiasi var a sbg b, nilai 7
b = tukar #inisiasi b sbg tukar, nilai 5
print('setelah tukar: ')
print('a=',a) #nilai a jd 7
print('b=',b) #nilai b jd 5

#menukar variable dgn one liner
a = 5
b = 7
a, b = b, a
print('setelah tukar: ')
print('a: ',a)
print('b: ',b)
```

```python
#PERCABANGAN DAN TERNARY OPERATORS
'''
- control flow adlh cara utk memberi tahu perogram mengenai instruksi yg hrs
dijalankan dan di mana hrs memulai dan berakhir.
- ctrl flow berjalan berdsrkan jalur eksekusi.'''
#PERCABANGAN

ada_haikyuu = "nonton movie" #klo dihilangkan, nnti outputnya yg statement "else"
if ada_haikyuu == "nonton movie":
  print("saya datang dan nonton haikyuu")
else:
  print("saya pulang dan mengsedih")

#STATEMENT IF
'''
utk mengecek apakah nilai variable di dlmnya memenuhi kriteria suatu kondisi
atau tidak. jika memenuhi, maka bernilai true. jika tdk, maka false
'''
#EXAMPLE 1
skor = 100
if skor ==100:
  print("Excellent!")

#EXAMPLE 1, ONE LINER VERSE
skor = 100
if skor == 100: print("Excellent!")

#STATEMENT ELSE
'''
digunakan saat mjd jalan keluar saat kondisi/hasil evaluasi if statement bernilai
false.'''

#EXAMPLE 2, IF AND ELSE STATEMENT
ada_duid = "nonton haikyuu"
if ada_duid == "nonton haikyuu":
  print("Silahkan menikmati Nekoma vs Karasuno")

else:
  print("Silahkan cari cuan lagi. Semangat!")
```

```python
#EXAMPLE 3, ELIF STATEMENT
'''
merup kependekan dari else dan if, alternatif utk if bertingkat atau switch case.
statement elif ditaruh setlh if, tdk ada batasan utk seberp byk.'''

#EXAMPLE 3: IF, ELIF, AND ELSE

IPK = 2.78

if IPK>=3.70:
  print("Selamat, kamu dapat predikat summa cumlaude!")
  print("Pertahankan!")

elif IPK>3.50:
  print("Selamat, kamu dapat predikat cumlaude!")
  print("Udah keren banget tuh..")

elif IPK>=3.30:
  print("Selamat, kamu dapat predikat sangat memuaskan!")
  print("Tingkatkan lagi.")

else:
  print("Terima kasih sudah bertahan!")
  print("Jangan menyerah, ya.")
```

```python
#GABUNG DENGAN OPERATOR
IPK = 3.78
perilaku = 'tidak baik'

if IPK>=3.50 and perilaku =='baik':
  print("Selamat, kamu dapat predikat cumlaude dan berkelakuan baik!")
  print("Pertahankan!")

elif IPK>=3.50 and perilaku !='baik':
  print("Selamat, kamu dapat predikat cumlaude, tapi perilaku kamu kurang baik.")
  print("Perbaiki lagi ya..")

else:
  print("Terima kasih sudah bertahan!")
  print("Jangan menyerah, ya.")

```

```python
#TERNARY OPERATORS
'''
termasuk conditional expression pd python, maksudnya btk ekspresi yg ber7an
utk mengevaluasi kondisi & mengembalikan hasil evaluasinya.
versi one liner dari if dan else.
#gua agk familiar, mirip kyk visual blocknya robockly sense## writer's note
'''
lulus = False
print("omedetou") if lulus else print("daijoubu")

#BLOCK CODE VERSE
lulus = False
if lulus:
  print("omedetou")
else:
  print("daijoubu")

#TERNARY OPERATORS WITH TUPLE, ONE LINER
lulus = True
graduation = ("Daijoubu, try again!", "Omedetou, sugoi neee")[lulus]
print(graduation)

#TERNARY OPERATORS WITH TUPLE, BLOCK CODE
lulus = False
if lulus:
  print("Omedetou, sugoi neee")
else:
  print("Daijoubu, try again!")
```
